```python
new_df = df[df.columns[df.isna().sum()!=0]]
new_df
```

```python
temp_df = new_df.copy()
temp_df
```

```python
for i in ["Cash Ratio","Current Ratio","Quick Ratio","Earnings Per Share","Estimated Shares Outstanding"]:  
    temp_df[i].fillna(temp_df[i].median(),inplace=True)
    
temp_df["For Year"].fillna(temp_df["For Year"].mode(),inplace=True)
```
